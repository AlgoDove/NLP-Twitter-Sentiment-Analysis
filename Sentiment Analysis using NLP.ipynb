```python
def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http.?://[^\s]+[\s]?', '', text)

    # Tokenize
    words = nltk.word_tokenize(text)

    # Remove stopwords and punctuation
    filtered_words = [
        word for word in words
        if word not in stopwords.words('english') and word not in string.punctuation
    ]

    # Apply stemming
    stemmed_words = [ps.stem(word) for word in filtered_words]

    # Join back to a string
    return " ".join(stemmed_words)
```

```python
df = pd.read_csv("Tweets.csv")  # Make sure the filename matches exactly
df = df[["airline_sentiment", "text"]]
df.head()

```

```python
import re
import string
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')

ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize by splitting on whitespace
    words = text.split()
    # Remove stopwords and stem
    cleaned_words = [ps.stem(word) for word in words if word not in stop_words]
    return " ".join(cleaned_words)
```

```python
df["text_cleaned"] = df["text"].apply(clean_text)
df[["text", "text_cleaned"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF vectorizer with max 3000 features
tfidf = TfidfVectorizer(max_features=3000)

# Fit and transform cleaned text into feature vectors
X = tfidf.fit_transform(df["text_cleaned"]).toarray()

# Target variable: sentiment labels
Y = df["airline_sentiment"].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)

y_pred_nb = nb_model.predict(X_test)
print("Naive Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```
